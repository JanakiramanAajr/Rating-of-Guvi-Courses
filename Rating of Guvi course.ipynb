```python
y  = df['Rating']
x = df.drop(['Rating','published_timestamp'],axis = 1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.15)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# sample algorithm for reference
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
result = model.fit(x_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

# Assuming you have your model and predictions
y_pred = result.predict(x_test)

# Calculate the Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)

# Calculate the Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)

# Calculate the R-squared (R2)
r2 = r2_score(y_test, y_pred)

print('Mean Squared Error (MSE):', mse)
print('Root Mean Squared Error (RMSE):', rmse)
print('R-squared (R2):', r2)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import (RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor,
                              ExtraTreesRegressor, HistGradientBoostingRegressor)
from sklearn.svm import SVR
reg_algo = {'DecisionTreeRegressor': DecisionTreeRegressor,
            'RandomForestRegressor': RandomForestRegressor,
            'AdaBoostRegressor': AdaBoostRegressor,
            'GradientBoostingRegressor': GradientBoostingRegressor,
            'ExtraTreesRegressor': ExtraTreesRegressor,
            'HistGradientBoostingRegressor':HistGradientBoostingRegressor,
            'SVR': SVR}
for key, value in reg_algo.items():
    model = value()
    result = model.fit(x_train, y_train)
    pickle.dump(result,open(key +'.sav','wb'))
    
    
    
```
